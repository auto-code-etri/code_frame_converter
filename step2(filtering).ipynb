```python
from json_tool import *
from filter_tool import *
```

### 기존의 jsonl 형태로 저장된 코드들을 데이터 프레임 형태로 변환한다.

```python
#jsonl_file = "code_source/Codes2_Codes3_base.jsonl"
jsonl_file = "code_source/Code_base_example.jsonl"
df = read_jsonl_file(jsonl_file)  # lines=True는 JSONL 파일용, False는 JSON 파일용
df
```

### 500 라인 이상인 코드는 모두 제거한다.

```python
lineno = 500
df2 = filter_by_numline_limit(df, lineno) # OO 라인 이하인 것은 전부 제거
df2
```

```python
minlineno = 3
df2 = filter_by_numline_over(df2, minlineno) # OO 라인 이하인것은 전부 제거거
df2
```

### 코드 중에서 C/C++ 함수 형태가 아닌 코드들이 있다. 이러한 코드들은 제거한다

```python
df3 = filter_by_non_cpp_functions(df2) # C/C++가 아닌 엔트리는 제거
df3
```

### 코드 중에서 파서의 오류로 함수가 1개 존재하는 것이 아니라 2개 이상 존재할 수 있다. 이러한 경우는 제거한다. (함수 1개만 허용)

```python
df4 = filter_by_removing_multiple_functions(df3) # 코드에 함수가 여러개 있는 것은 제거
df4
```

### 코드의 라인 수 뿐만 아니라 토큰 수도 제한을 둔다. 5000개의 토근 이상인 경우, LLM이 처리하기에는 부담스러우므로 제거한다.

```python
tokenno = 5000
df5 = filter_by_numtoken_limit(df4, tokenno)
df5
```

### 필터된 jsonl 파일을 저장한다.

```python
output_jsonl_file = "code_filtered/Code_filtered_example.jsonl"
convert_df_to_jsonl(df5, output_jsonl_file)
```
